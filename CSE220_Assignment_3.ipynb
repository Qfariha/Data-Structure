```python
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
import traitlets
class LinkedList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array. head will refer
  #  to the Node that contains the element from a[0]
  #  Else Sets the value of head. head will refer
  # to the given LinkedList

  # Hint: Use the type() function to determine the data type of a
    if type(a)==list:
      self.list1=a
      self.head = Node(a[0],None)
      tail=self.head
    
      for i in range(1,len(self.list1)):
        n=Node(a[i],None)  #next obj creation
        tail.next=n        #next er info store
        tail=tail.next     #last elem loc update
    
    else:
      self.head=a
    
  # Count the number of nodes in the list
  def countNode(self):
    # To Do
    temp=self.head
    count=0
    while temp!=None: #if temp.next!=None, loop ek ghor kom chole karon last elem er next a None dekhe last elem skip kore dey
      count+=1
      temp=temp.next
    return count
  
  # Print elements in the list
  def printList(self):
    # To Do
    temp=self.head
    while temp!=None:
      print(temp.element,end=", ")
      temp=temp.next
    print()
  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    #pass # Remove this line
    temp=self.head
    count=self.countNode()
    index=0
    
    if idx<0 or idx>count-1: #count-1=last index
      return None
    
    while temp!=None:  #else dei nai karon if a return korle ar code run korbe na
      if index==idx:
        return temp
      index+=1
      temp=temp.next
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    # To Do
    temp=self.head
    count=self.countNode()
    index=0
    
    if idx<0 or idx>count-1: #count-1=last index
      return None
    
    while temp!=None:  #else dei nai karon if a return korle ar code run korbe na
      if index==idx:
        return temp.element
      index+=1
      temp=temp.next
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    # To Do
    temp=self.head
    count=self.countNode()
    index=0
    
    if idx<0 or idx>count-1:
      return None
    
    while temp!=None:  
      if index==idx:
        temp.element=elem
        return temp.element
      index+=1
      temp=temp.next
  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp=self.head
    count=self.countNode()
    index=0
    if elem not in self.list1:

      return -1
    while temp!=None:  
      if temp.element==elem:
        
        return index
      index+=1
      temp=temp.next
    
  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    # To Do
    temp=self.head
    count=self.countNode()
    
    while temp!=None:  
      if elem not in temp.element:
    
        return False
      temp=temp.next
      return True

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    # To Do
    nh=Node(self.head.element,None)  #newlist
    tail=nh
    temp=self.head.next
    while temp!=None:
      n=Node(temp.element,None)
      tail.next=n
      tail=tail.next
      temp=temp.next  #temp pore increase korbo naile node skip hoye jabe
    return nh

  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    #To Do
    rh=Node(self.head.element,None)
    temp=self.head.next
    while temp!=None:
      n=Node(temp.element,rh)
      rh=n
      temp=temp.next
    return rh
                                             #    rh
    # rh=None                                  # n= #10,20,30,40
    # n=self.head                              #     n  nn
    # while n!=None:
    #   nextnode=n.next                          #nextnode= #20,30,40 
    #   n.next=rh                               #n.next= None
    #   rh=n                                    #rh=10
    #   n=nextnode                               #n=20
  
  #10 20 30 40--------> 40->30->20->10

  # inserts Node containing the given element at the given index
  # Check validity of index.
  def insert(self, elem, idx):
    # To Do
    count=self.countNode()
    if idx>count or idx<0:  
      return "Invalid index"
    elif idx==0:
      n=Node(elem,self.head)
      n.next=self.head  
      self.head=n    #or else head will be lost
    elif idx==count: #last idx er pore
      x=self.nodeAt(idx-1)  #last node  
      n=Node(elem,None)
      x.next=n
    else:
      n=Node(elem,None)
      n1=self.nodeAt(idx)  #current node at that index
      n.next=n1 #new node a current node er shathe connection
      n_prev=self.nodeAt(idx-1)
      n_prev.next=n
    
    


  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    count=self.countNode()
    removed_elem=self.get(idx)
    if idx>count-1 or idx<0:
      return "Invalid index"
    elif idx==0:
      temp=self.head
      self.head=temp.next
      return removed_elem
    elif idx==count-1:  #last idx
      x=self.nodeAt(idx-1)  #last er ager node
      x.next=None
      return removed_elem
    else:
      n_prev=self.nodeAt(idx-1)
      n_next=self.nodeAt(idx+1)
      n_prev.next=n_next
    
      return removed_elem  
      
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):

    

    # self.insert(self.head.element,count)
    # self.remove(0)
    
    temp=self.head  #10,20,30,40
    self.head=temp.next  #20,30,40
  
    count=self.countNode()  #3
    last_node=self.nodeAt(count-1)  #3-1=2, l_n=40
    temp.next=None       #temp=10,20,30,40 ,temp.next=20,30,40 so temp=10 only
    last_node.next=temp #20,30,40,10
    


  
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    
    temp1=self.head     #10,20,30,40
    temp2=temp1
    sl_node=self.nodeAt(self.countNode()-2)  #2nd last node

    while temp1!=None:
      temp3=temp1       #loop agaite thakbe tai sheshe temp3 te last node thakbe  
                         #temp3=40
      temp1=temp1.next 

    temp3.next=temp2  #last node er next a head add korlam
                      #40,10,20,30,40
    self.head=temp3
    sl_node.next=None   #2nd last node er next None kore dilam #30 er next=None
    
    return self.head
   
```
