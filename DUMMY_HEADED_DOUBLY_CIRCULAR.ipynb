```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array.
    self.head = Node(None, None, None)
    #self.head.next = self.head.prev = self.head
    if type(a) == list:
      tail = self.head
      for i in range(len(a)):
        node = Node(a[i], None, None)
        node.prev = tail
        # node.next = self.head
        tail.next = node
        # self.head.prev = node
        # node.prev = tail
        tail = node
      tail.next = self.head
      self.head.prev = tail
    
    else:
      self.head=a
  
  # Counts the number of Nodes in the list
  def countNode(self):
    
    temp=self.head.next
    count=0
    while temp!=self.head:
      count+=1
      temp=temp.next
    return count

    print()
  # prints the elements in the list
  def forwardprint(self):
    
    temp=self.head.next
    while temp!=self.head:  #temp.next dile agei next a none peye jaay tai loop ek step kom chole
      print(temp.element,end=", ")
      temp=temp.next

    print()
  # prints the elements in the list backward
  def backwardprint(self):

    temp=self.head.next

    while temp.next!=self.head: #loop ek ghor kom cholbe last elem porjonto jawar jonno #just like len-1,i+=1
      temp=temp.next       
    while temp!=self.head:   #last index a giye erpor back kortesi  
                        # N,1,2,3,N #temp.next!=None korle 1 a eshe loop shesh hoye jaito
                            
      print(temp.element,end=", ")
      temp=temp.prev
    print()
  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    temp=self.head.next
    count=self.countNode()
    index=0
    
    if idx<0 or idx>count-1:  #count-1 cause last index theke boro hole
      return "Index Error"
    
    while temp!=self.head:  #else dei nai karon if a return korle ar code run korbe na
      if index==idx:
        return temp
      index+=1
      temp=temp.next
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp=self.head.next
    count=self.countNode()
    index=0
    
    while temp!=self.head:  
      if temp.element==elem:
        
        return index
      index+=1
      temp=temp.next
    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    count=self.countNode()
    if idx>count or idx<0:
      return "Invalid index"
    elif idx==0:
      temp = self.head.next
      node = Node(elem, None, None)
      node.next = temp
      node.prev = self.head
      temp.prev = node
      self.head.next = node
      temp = node
    elif idx==count: #last idx
      temp = self.nodeAt(idx-1)
      node = Node(elem, None, None)
      temp.next = node
      node.prev = temp
      node.next = self.head
      self.head.prev = node
      temp = node
      
    else:
      n=Node(elem,None,None)
      n1=self.nodeAt(idx)  #current node at that index
      
      n1.prev=n
      n.next=n1
      n_prev=self.nodeAt(idx-1)
      n_prev.next=n
      n.prev=n_prev
      
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.

  def remove(self, idx):
    count=self.countNode()
    
    if idx>count or idx<0:
      return "Invalid index"
    elif idx == 0:
      temp = self.head.next
      self.head.next = temp.next
      temp.next.prev = self.head
    else: 
      temp = self.nodeAt(idx)
      tail = self.nodeAt(idx-1)
      tail.next = temp.next
      temp.next.prev = tail
      tail = tail.next.next
    return str(temp.element)

  #extraaaaaaaaaaaaaaaa
  # def reverse(self,a):
  #   a_head=a.head
  #   rh=Node(a_head.element,None,None)
  #   tem=a_head.next
  #   while tem.next!=None:
  #     n=Node(a_head.next.element,rh,None)    #1,2,3,4---> N<-2-><-1-N , 
  #     rh.prev=n
  #     rh.next=n.prev
  #     rh=n
  #     tem=tem.next
  #   return rh
  # def reverse(self,a):
  #   head=a.head
  #   tail=a.head
  #   count=0
  #   while tail.next!=None:
  #     tail=tail.next
  #     count+=1
    
  #   for i in range(count//2):
      
  #     temp=head.element
  #     head.element=tail.element
  #     tail.element=tail
      
  #     tail=tail.prev
  #     head=head.next

  #   return head
    
```
