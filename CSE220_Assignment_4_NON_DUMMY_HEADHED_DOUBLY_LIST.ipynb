```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array.
    if type(a)==list:
      self.list1=a
      self.head = Node(a[0],None,None)
      tail=self.head
    
      for i in range(1,len(self.list1)):
        n=Node(a[i],None,None)  #next obj creation
        tail.next=n        #next er info store
        n.prev=tail     #next node a connection dewa prev er 
        tail=tail.next   #tail update
        # tail.next=self.head  #for circular
        # self.head.prev=tail
    
    else:
      self.head=a
  
  # Counts the number of Nodes in the list
  def countNode(self):
    
    temp=self.head
    count=0
    while temp!=None:
      count+=1
      temp=temp.next
    return count

    print()
  # prints the elements in the list
  def forwardprint(self):
    
    temp=self.head
    while temp!=None:  #temp.next dile agei next a none peye jaay tai loop ek step kom chole
      print(temp.element,end=", ")
      temp=temp.next

    print()
  # prints the elements in the list backward
  def backwardprint(self):

    temp=self.head

    while temp!=None: #loop ek ghor kom cholbe last elem porjonto jawar jonno #just like len-1,i+=1
      temp2=temp
      temp=temp.next       
    while temp2!=None:   #last index a giye erpor back kortesi  
                        # N,1,2,3,N #temp.next!=None korle 1 a eshe loop shesh hoye jaito
                            
      print(temp2.element,end=", ")
      temp2=temp2.prev
    print()
  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    temp=self.head
    count=self.countNode()
    index=0
    
    if idx<0 or idx>count-1:  #count-1 cause last index theke boro hole
      return "Index Error"
    
    while temp!=None:  #else dei nai karon if a return korle ar code run korbe na
      if index==idx:
        return temp
      index+=1
      temp=temp.next
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp=self.head
    count=self.countNode()
    index=0
    if elem not in self.list1:

      return -1
    while temp!=None:  
      if temp.element==elem:
        
        return index
      index+=1
      temp=temp.next
    return None

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    count=self.countNode()
    if idx>count or idx<0:
      return "Invalid index"
    elif idx==0:
      n=Node(elem,self.head,None)
      n.next=self.head  #or else head will be lost
      self.head.prev=n
      self.head=n
    elif idx==count: #last idx
      x=self.nodeAt(idx-1)  #last node  
      n=Node(elem,None,x)
      x.next=n
      n.prev=x
      
    else:
      n=Node(elem,None,None)
      n1=self.nodeAt(idx)  #current node at that index
      
      n1.prev=n
      n_prev=self.nodeAt(idx-1)
      n_prev.next=n
      n.prev=n_prev
      
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.

  def remove(self, idx):
    count=self.countNode()
    
    if idx>count or idx<0:
      return "Invalid index"
    elif idx==0:
      elem=self.head.element
      temp=self.head
      self.head=temp.next
      self.head.prev=None
      return str(elem)
    elif idx==count-1:
      elem=self.nodeAt(idx) 
      x=self.nodeAt(idx-1)  #current
      x.next=None
      return str(elem)
    else:
      elem=self.nodeAt(idx) 
      n_prev=self.nodeAt(idx-1)
      n_next=self.nodeAt(idx+1)
      n_prev.next=n_next
      n_next.prev=n_prev
    
      return str(elem)  


  #extraaaaaaaaaaaaaaaa
  # def reverse(self,a):
  #   a_head=a.head
  #   rh=Node(a_head.element,None,None)
  #   tem=a_head.next
  #   while tem.next!=None:
  #     n=Node(a_head.next.element,rh,None)    #1,2,3,4---> N<-2-><-1-N , 
  #     rh.prev=n
  #     rh.next=n.prev
  #     rh=n
  #     tem=tem.next
  #   return rh
  def reverse(self,a):
    head=a.head
    tail=a.head
    count=0
    while tail.next!=None:
      tail=tail.next
      count+=1
    
    for i in range(count//2):
      
      temp=head.element
      head.element=tail.element
      tail.element=tail
      
      tail=tail.prev
      head=head.next

    return head
    
```
